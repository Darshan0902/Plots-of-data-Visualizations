# Libraries used and their definitons ...

<h2> Matplotlib.pyplot </h2> 

matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

<h2> numpy </h2>

NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.

```python
#importing basic libraries essential for our code...

import matplotlib.pyplot as plt #importing matplot library using ".pyplot" used for making graphs and plots
import numpy as np # numpy is used for arrays and operations to be performed on them .
```

```python
# Randomly sampled data
x = np.random.normal(0, 1, 1000)  # The code is used to generate a random number between 0 and 1,000 for x 
y = np.random.normal(0, 1, 1000)  # The code is used to generate a random number between 0 and 1,000 for y.
z = np.random.normal(0, 1, 1000)  #  The code is used to generate a random number between 0 and 1,000 for z.


labels = ['A', 'B', 'C', 'D', 'E'] # A list having a set of strings.
values = [25, 35, 28, 9 , 3] # A list having a set of Integers.
```

A histogram is a graphical representation of data points organized into user-specified ranges. Similar in appearance to a bar graph, the histogram condenses a data series into an easily interpreted visual by taking many data points and grouping them into logical ranges or bins .  A histogram displays numerical data by grouping data into "bins" of equal width. Each bin is plotted as a bar whose height corresponds to how many data points are in that bin.Bins are also sometimes called "intervals", "classes", or "buckets".


```python
# Creating a Histogram 

# plt.hist() is a function of pyplot used to create a Histogram 
plt.hist(x, bins=30)

#setting the x label as "value"
plt.xlabel('Value') 

#setting the y label to be "Frequency".
plt.ylabel('Frequency') 

#setting the title of the graph as "Histogram"
plt.title('Histogram') 

#plt.show is used to show the plotted functions above..
plt.show()
```

A  bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value. Some bar graphs present bars clustered in groups of more than one, showing the values of more than one measured variable

```python
# Creating a Barchart :

# plt.bar() is a function used to draw a barchart using pyplot module of matplotlib.
# Here we decided the X_axis to be "labels" and the y_axis to be "values"
plt.bar(labels, values)

#Labelling the title of x-axis as "Category".
plt.xlabel('Category')

#Labelling the y axis as "Value".
plt.ylabel('Value')

# Setting the title of the below graph to be "Bar Chart"
plt.title('Bar Chart')

# The same plt.show function used to display your graph or plot function...
plt.show()

```

A pie chart, sometimes called a circle chart, is a way of summarizing a set of nominal data or displaying the different values of a given variable (e.g. percentage distribution). This type of chart is a circle divided into a series of segments. Each segment represents a particular category

```python
# A pie chart is created using plt.pie() function in pyplot.
# In the below statetment we declared values from the list of "values" and choses the list of "labels" to label them.
plt.pie(values, labels=labels)

# We use plt.title to set the label of any particular graph..
plt.title('A Pie or a Circular Chart')

# plt.show is used for displaying the graph.
plt.show()
```

# BOX-PLOT :

<p style="padding: 10px; background-color: #000000; margin: 0; color: ORANGE; font-family: New Times Roman; font-size: 135%; text-align: center; border-radius: 5px; overflow: hidden; font-weight: 500;">
    BOX-PLOT
</p>

<p style="text-align: center;">
    <!-- Content related to the bar chart can be added here -->
</p>


A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
# plt.boxplot() is a function used to initialize a boxplot with three variables as x and y
plt.boxplot([x, y])

# Labelling X-axis to be "Variable".
plt.xlabel('Variable')

# Labelling the Y-axis as "Value"
plt.ylabel('Value')

#Label the title of graph as "Box Plot".
plt.title('Box Plot')

#Last but not Least the plt.show() to display your graph
plt.show()

```

# Scatterplot:

<p style="padding: 10px; background-color: #000000; margin: 0; color: ORANGE; font-family: New Times Roman; font-size: 135%; text-align: center; border-radius: 5px; overflow: hidden; font-weight: 500;">
    SCATTERPLOT
</p>

<p style="text-align: center;">
    <!-- Content related to the bar chart can be added here -->
</p>

Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis. These plots are often called scatter graphs or scatter diagrams.



```python
# use plt.scatter( ) to display a scatterplot.
plt.scatter(x, y)

# Plot the label of X-axis as "X".
plt.xlabel('X')

# Plot the label of Y-axis as "Y"
plt.ylabel('Y')

#Label the Title of graph as "Scatter Plot".
plt.title('Scatter Plot')

# Using plt.show() to display the graph AGAIN...
plt.show()
```

# Violin Plot :

<p style="padding: 10px; background-color: #000000; margin: 0; color: ORANGE; font-family: New Times Roman; font-size: 135%; text-align: center; border-radius: 5px; overflow: hidden; font-weight: 500;">
    VIOLIN PLOT
</p>

<p style="text-align: center;">
    <!-- Content related to the bar chart can be added here -->
</p>


A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.



```python
# plt.violinplot() is used to plot a Violin plot . Here with two variables [x,y]
plt.violinplot([x, y])

# Set the label of X-axis as "Variable"
plt.xlabel('Variable')

#Set the label of Y-axis to be "Value"
plt.ylabel('Value')

# Set the title of the Graph as "Violin Plot"
plt.title('Violin Plot')

# plt.show is used to display the Graph.
plt.show()
```

# WORD CLOUD : 


<p style="padding: 10px; background-color: #000000; margin: 0; color: ORANGE; font-family: New Times Roman; font-size: 135%; text-align: center; border-radius: 5px; overflow: hidden; font-weight: 500;">
WORD CLOUD 
</p>

<p style="text-align: center;">
    <!-- Content related to the bar chart can be added here -->
</p>



A word cloud is a visual representation of text data where the size of each word corresponds to its frequency or importance within the text. It is a popular data visualization technique used to quickly grasp the key themes, keywords, or trends present in a body of text. Word clouds provide an intuitive and visually appealing way to highlight the most prominent words in a text corpus.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Sample data (replace with your own data)
text = "data analytics visualization machine learning python kaggle data science"

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

# Create and configure the plot
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")

# Add a title
plt.title("Word Cloud")

# Display the plot
plt.show()
```
